```python
## Registro civil totem Autoservicio ##
from IPython.display import clear_output
from time import sleep


tabla_datos = [(),(),(),(),()]

repetición = 0;

def verificar_rut(r,n):
  rut = str(r)
  reverse_rut = rut[::-1]
  factor = 2
  suma = 0

  for digito in reverse_rut:
    suma+= int(digito) * factor
    factor += 1
    if factor > 7:
      factor = 2

  digito_verificador = 11 - (suma % 11)
  if digito_verificador == 11:
    digito_verificador = "0"
  elif digito_verificador == 10:
    digito_verificador = "k"

  if len(r) == len("xxxxxxxx"):
    if str(digito_verificador) == n:
        return True
      elif str(digito_verificador) != n:
        return False
  elif len(r) != len("xxxxxxxx"):
    return False


def grabar_datos(r,n,e,p,c): #primera función graba los datos dentro de la lista
  tabla_datos[0] = list(tabla_datos[0])
  tabla_datos[0].append(r)
  tabla_datos[0] = tuple(tabla_datos[0])
  if len(n) >= len("xxxxxxxx"):
    tabla_datos[1] = list(tabla_datos[1])
    tabla_datos[1].append(n)
    tabla_datos[1] = tuple(tabla_datos[1])
    if e > 0:
      tabla_datos[2] = list(tabla_datos[2])
      tabla_datos[2].append(e)
      tabla_datos[2] = tuple(tabla_datos[2])
      if type(p) is str and type(c) is str:
        tabla_datos[3] = list(tabla_datos[3])
        tabla_datos[3].append(p)
        tabla_datos[3] = tuple(tabla_datos[3])
        tabla_datos[4] = list(tabla_datos[4])
        tabla_datos[4].append(c)
        tabla_datos[4] = tuple(tabla_datos[4])
        return True
      else:
        return 2
  else:
    return 2

def comprobar(r):#comprueba si dentro de la lista ya existe el rut puesto
    if r in tabla_datos[0]:
        return False
    else:
        return True
def Tabla():#imprime una tabla con los datos ingresado
    for a in range(len(tabla_datos[0])):
        print("\n",tabla_datos[0][a],
            tabla_datos[1][a],
            tabla_datos[2][a],
            tabla_datos[3][a],
            tabla_datos[4][a])
def ruts():#imprime cada uno de los rut ingresado
    for i in tabla_datos[0]:
        print(f"\n{i}")
def buscar(b):#busca un rut especifico dentro de la lista
    for a in range(len(tabla_datos[0])):
        if b in tabla_datos[0][a]:
            return tabla_datos[0][a], tabla_datos[1][a], tabla_datos[2][a], tabla_datos[3][a], tabla_datos[4][a]
def documento(t,rut):#crea tres distintos tipos de documentos
    for a in range(len(tabla_datos[0])):
        if rut in tabla_datos[0][a]:
            if t == 1:
              clear_output()
              r = tabla_datos[0][a]
              n = tabla_datos[1][a]
              e = tabla_datos[2][a]
              p = tabla_datos[3][a]
              c = tabla_datos[4][a]
              print(" ------------------------------"
                    "\n|         Certificado          |"
                    "\n|              de              |"
                    "\n|          nacimiento          |"
                    "\n|                              |"
                    f"\n| rut:{r}               |"
                    f"\n| nombre:{n}            |"
                    f"\n| edad:{e}                      |"
                    f"\n| país:{p}                   |"
                    f"\n| ciudad:{c}              |"
                    "\n|                              |"
                    "\n|                              |"
                    "\n ------------------------------"
                    )
            if t == 2:
              clear_output()
              r = tabla_datos[0][a]
              n = tabla_datos[1][a]
              e = tabla_datos[2][a]
              p = tabla_datos[3][a]
              c = tabla_datos[4][a]
              print(" ------------------------------"
                    "\n|         Certificado          |"
                    "\n|           conyugal           |"
                    "\n|                              |"
                    "\n|                              |"
                    f"\n| rut:{r}               |"
                    f"\n| nombre:{n}            |"
                    f"\n| edad:{e}                      |"
                    f"\n| país:{p}                   |"
                    f"\n| ciudad:{c}              |"
                    "\n|                              |"
                    "\n|                              |"
                    "\n|                              |"
                    "\n ------------------------------"
                    )
            if t == 3:
              clear_output
              r = tabla_datos[0][a]
              n = tabla_datos[1][a]
              e = tabla_datos[2][a]
              p = tabla_datos[3][a]
              c = tabla_datos[4][a]
              print(" ------------------------------"
                    "\n|         Certificado          |"
                    "\n|              de              |"
                    "\n|         pertenencia          |"
                    "\n|                              |"
                    f"\n| rut:{r}               |"
                    f"\n| nombre:{n}            |"
                    f"\n| edad:{e}                      |"
                    f"\n| país:{p}                   |"
                    f"\n| ciudad:{c}              |"
                    "\n|                              |"
                    "\n|                              |"
                    "\n ------------------------------"
                    )
def eliminar(e):#elimina los elementos relacionado a un rut
  for a in range(len(tabla_datos[0])):
      if e in tabla_datos[0][a]:
        tabla_datos[0] = list(tabla_datos[0])
        del tabla_datos[0][a]
        tabla_datos[0] = tuple(tabla_datos[0])

        tabla_datos[1] = list(tabla_datos[1])
        del tabla_datos[1][a]
        tabla_datos[1] = tuple(tabla_datos[1])

        tabla_datos[2] = list(tabla_datos[2])
        del tabla_datos[2][a]
        tabla_datos[2] = tuple(tabla_datos[2])

        tabla_datos[3] = list(tabla_datos[3])
        del tabla_datos[3][a]
        tabla_datos[3] = tuple(tabla_datos[3])

        tabla_datos[4] = list(tabla_datos[4])
        del tabla_datos[4][a]
        tabla_datos[4] = tuple(tabla_datos[4])
        break

while True:#menu para hacer funcionar el servicio
    clear_output()
    print("Bienvenido a autoservicio civil, ¿que desea hacer?")
    print(
        "\n\t1- Grabar Datos"
        "\n\t2- Buscar Datos"
        "\n\t3- Imprimir Certificados"
        "\n\t4- Eliminar Datos"
        "\n\t5- Consultar Datos"
        "\n\t6- Salir del menu"
        )
    choise = int(input("\npresione el número correspondiente a lo que desea hacer: "))
    try:
        if choise == 1:
            repetición += 1
            clear_output()
            rut = input("\ningrese rut sin puntos: \t\t")
            NV = input("ingrese numero verificador de su rut: \t")
            nombre = input("\ningrese nombre:\t\t\t\t")
            edad = int(input("ingrese edad:\t\t\t\t"))
            país = str(input("ingrese país de nacimiento:\t\t"))
            ciudad = str(input("ingrese ciudad de nacimiento:\t\t"))
            if comprobar(rut) is False:
                clear_output()
                print("\nDatos ya ingresados, ¡¡ADVERTENCIA: el rut no se puede repetir!!")
            elif comprobar(rut) is True:
              if verificar_rut(rut, NV) is True:
                c_rut = rut + "-" + NV
                u = grabar_datos(c_rut,nombre,edad,país,ciudad)
                if u is False:
                  clear_output()
                  print("\nlos datos ingresados no son validos")
                  if repetición >= 1:
                      e = 0
                      tabla_datos[0] = list(tabla_datos[0])
                      del tabla_datos[0][2]
                      tabla_datos[0] = tuple(tabla_datos[0])
                      tabla_datos[1] = list(tabla_datos[1])
                      del tabla_datos[1][2]
                      tabla_datos[1] = tuple(tabla_datos[1])
                      tabla_datos[2] = list(tabla_datos[2])
                      del tabla_datos[2][2]
                      tabla_datos[2] = tuple(tabla_datos[2])
                      e +=1
                      sleep(2)
                      repetición -= 1
                elif u == 2:
                    clear_output()
                    print("\nlos datos ingresados no son validos")
                    sleep(2)
                    if repetición >= 1:
                      e = 0
                      tabla_datos[0] = list(tabla_datos[0])
                      del tabla_datos[0][2]
                      tabla_datos[0] = tuple(tabla_datos[0])
                      tabla_datos[1] = list(tabla_datos[1])
                      del tabla_datos[1][2]
                      tabla_datos[1] = tuple(tabla_datos[1])
                      tabla_datos[2] = list(tabla_datos[2])
                      del tabla_datos[2][2]
                      tabla_datos[2] = tuple(tabla_datos[2])
                      e +=1
                elif u is True:
                    clear_output()
                    sleep(3)
                    print("\ndatos guardados correctamente")
              elif verificar_rut(rut,NV) is False:
                  print("el rut ingresado no es valido")
                  sleep(2)
            salir = int(input("\ndesea continuar 1-y/2-n\t"))
            if salir == 1:
                clear_output()
                print("\nvolviendo...")
                sleep(3)
            elif salir == 2:
                clear_output()
                print("Usted ha salido con éxito")
                break
        if choise == 2:
            if len(tabla_datos[0]) > 0:
                clear_output()
                print("\nlos datos ingresados son: \n")
                ruts()
                busca = input("\ningrese el rut que desea buscar: \n")
                bu = buscar(busca)
                print(f"\n{bu}")
                sleep(3)
                clear_output()
                print("\nvolviendo...")
                sleep(3)
            else:
                clear_output()
                print("\naun no se han ingresado datos")
                sleep(2)
                clear_output()
                print("\nvolviendo...")
                sleep(3)
        if choise == 3:
            clear_output()
            if len(tabla_datos[0]) > 0:
                clear_output
                ruts()
                rut = input("ingrese el rut a buscar: ")
                if rut in tabla_datos[0]:
                    clear_output()
                    print("\n1-Certificado de nacimiento\n2-Estado conyugal\n3-pertenencia al país de Chile")
                    tipoD = int(input("\nseleccione el tipo de documento que desea: "))
                    documento(tipoD,rut)
                    print("certificado emitido")
                    sleep(3)
                    clear_output()
                    print("\nvolviendo...")
                    sleep(3)
                else:
                    print("el rut ingresado no esta registrado")
                    sleep(2)
                    clear_output()
                    print("\nvolviendo...")
                    sleep(3)
            else:
                print("aun no se ha ingresado datos")
                sleep(2)
                clear_output()
                print("\nvolviendo...")
                sleep(3)
        if choise == 4:
            clear_output()
            if len(tabla_datos[0]) > 0:
                print("\nlos datos ingresados son: \n")
                ruts()
                elimina = input("\ningrese los datos del rut que desea eliminar: ")
                eliminar(elimina)
                print("datos eliminados satisfactoriamente")
                sleep(2)
                clear_output()
                print("\nvolviendo...")
                sleep(3)
            else:
                print("\naun no se ha ingresado datos")
                sleep(2)
                clear_output()
                print("\nvolviendo...")
                sleep(3)
        if choise == 5:
            clear_output()
            if len(tabla_datos[0]) > 0:
                Tabla()
                sleep(3)
                clear_output()
                print("\nvolviendo...")
                sleep(3)
            else:
                clear_output()
                print("\naun no se han ingresado datos")
                sleep(2)
                clear_output()
                print("\nvolviendo...")
                sleep(3)
        if choise == 6:
            clear_output()
            print("Saliendo...")
            sleep(2)
            print("usted ha salido con éxito")
            break
        if choise == 29:
          print("\nESTER EGG")
          sleep(3)
          print("¡¡¡PATATA!!!")
          sleep(20)
          clear_output()
          print("\nvolviendo...")
          sleep(3)
    except ValueError:
        print("Datos inválidos")
```
